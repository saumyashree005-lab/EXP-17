```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = {
    'Days' : ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours' : [2,3,2.5,4,3.5],
    'Marks' : [50,60,55,70,65],
    'Attendance' : [80,90,85,95,90],
    'Sleep_Hours' : [6,7,6.5,8,7.5],
    'Assignments_Completed' : [10,12,11,13,12]
}
df = pd.DataFrame(data)
df
```

```python
#1. Line Chart (Trend)
plt.plot(df['Days'], df['Study_Hours'], label='Study Hours')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Saumya')
plt.legend()
plt.show()
```

```python
#Advanced Line CHart
plt.figure(figsize=(7,4))
plt.plot(df['Days'], df['Study_Hours'], marker='o', color = 'blue', label = 'Study Hours')
plt.plot(df['Days'], df['Marks'], marker='s', color = 'red', label = 'Marks')
plt.xlabel('Days')
plt.ylabel('Values')
plt.title('Saumya')
plt.legend()
plt.show()
```

```python
#Bar Chart
plt.bar(df['Days'], df['Marks'], color = 'teal')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.title('Saumya')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Original data for df, restored for this cell
data = {
    'Days' : ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours' : [2,3,2.5,4,3.5],
    'Marks' : [50,60,55,70,65],
    'Attendance' : [80,90,85,95,90],
    'Sleep_Hours' : [6,7,6.5,8,7.5],
    'Assignments_Completed' : [10,12,11,13,12]
}
df = pd.DataFrame(data)

#Advanced Bar Graph
bars = plt.bar(df['Days'], df['Marks'], color = 'lavender')
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y, str(y), ha = 'center', va = 'bottom')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.title('Saumya')
plt.grid(axis='y')
plt.show()
```

```python
# Histogram Distributions
# Show how values are spread
data = {
    'Marks' : [50,60,55,70,65,75,50,85,90,45,68,72,77,82,88]
}
df = pd.DataFrame(data)
plt.hist(df['Marks'], bins = 3, edgecolor = 'black', alpha = 0.30)
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Saumya')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#Example data (more values for better histogram)
marks = [50,60,55,70,55,70,65,75,80,85,90,45,68,72,72,82,88]
#Calculate mean
mean = np.mean(marks)
plt.figure(figsize=(7,4))
plt.hist(marks, bins = 5, color = 'skyblue', edgecolor = 'black', alpha = 0.30)
#Mean line
plt.axvline(mean, color = 'red', linestyle = 'dashed', linewidth = 2, label = f'Mean: {mean:.2f}')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Saumya')
plt.legend()
plt.grid(axis='y', linewidth =2,alpha=0.3)
plt.show()
```

```python
#Scatter Plot (Relationship)
import matplotlib.pyplot as plt
plt.scatter(df['Study_Hours'], df['Marks'])
plt.xlabel('Study_Hours')
plt.ylabel('Marks')
plt.title('Saumya')
plt.show()
```

```python
# Conditional Scatter Plot
# Example Category
df['Result'] = ['Fail', 'Pass', 'Fail', 'Pass', 'Pass']
colors = ['red' if result == 'Fail' else 'green' for result in df['Result']]
plt.scatter(df['Study_Hours'], df['Marks'], c = colors)
plt.xlabel('Study_Hours')
plt.ylabel('Marks')
plt.title('Saumya')
plt.show()
```
